<h1>Concatenacion</h2>

```python
dt1 = pd.read_csv('dataObtenida/test3.csv')

print(dt1.shape)
dt1.head()
```

```python
dt2 = pd.read_csv('dataObtenida/test2.csv')

print(dt2.shape)
dt2.head()
```

```python
dt3 = pd.concat([dt1,dt2])
print(dt3.shape)
dt3.head()
```

```python
dt3 = dt3.sort_values(by='Provincia')
```

```python
dt3['Provincia'] = dt3['Provincia'].str.lower()
dt3['Localidad'] = dt3['Localidad'].str.lower()
dt3['calle_avenida'] = dt3['calle_avenida'].str.lower()
```

```python
dt3['Provincia'] = dt3['Provincia'].str.replace('entre ríos', 'entre rios')
dt3['Provincia'] = dt3['Provincia'].str.replace('tucumán', 'tucuman')
dt3['Provincia'] = dt3['Provincia'].str.replace('córdoba', 'cordoba')
dt3['Provincia'] = dt3['Provincia'].str.replace('río negro', 'rio negro')
dt3['Provincia'] = dt3['Provincia'].str.replace('santa fé', 'santa fe')
```

```python
dt3.drop(dt3[dt3['Provincia'].isna()].index , inplace=True)
dt3.shape
```

```python
dt3.to_csv('dataObtenida/CPA.csv', index=False)
```
